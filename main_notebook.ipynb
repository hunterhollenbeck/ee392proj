```python

import csv
import itertools
import matplotlib.pyplot as plt
```

```python
def harvest_data(filename,x,y,z):
    values = []

    with open(filename) as file_object:
        for line in csv.reader(file_object, delimiter=','):
            values.append(list(map(float, line)))
    for item in values:
        x.append(item[0])
        y.append(item[1])
        z.append(item[2])
```

```python


xs = []
ys = []
zs = []

harvest_data('rotation_test.txt', xs, ys, zs)
```
